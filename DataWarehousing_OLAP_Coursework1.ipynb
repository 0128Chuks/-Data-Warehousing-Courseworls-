# **Materialised view to answer Q1 and Q2**


*   **Q1. For each venue and year report the vaccination rates per vaccine, quarantine and drug treatment numbers.**

*  **Q2. For each year report the quarantine and drug treatment numbers.**



```python
# create Materialised View M1 in SQLITE, creating a window to reference later for analysis

cur.execute('''CREATE VIEW M1_Q1_Q2  AS
SELECT
    Year,
    venue_name,
    vaccine_name,
    SUM(drug_treatment) AS drug_treatment,
    SUM(QuarantineTotal) AS QuarantineTotal,
    SUM(VaccineCount) AS VaccineCount
FROM (
    SELECT
        0 AS drug_treatment,
        COUNT(FactQuarantine.citizenid) AS QuarantineTotal,
        0 AS VaccineCount,
        COALESCE(dimTime.Yearid, 'UNKNOWN') AS Year,
        COALESCE(dimVenue.venue_name, 'UNKNOWN') AS venue_name,
        'NONE' AS vaccine_name
    FROM FactQuarantine
    LEFT JOIN dimCitizen ON FactQuarantine.citizenid = dimCitizen.citizenid
    LEFT JOIN dimTime ON FactQuarantine.dateid = dimTime.dateid
    LEFT JOIN dimVenue ON FactQuarantine.venueid = dimVenue.venueid
    GROUP BY dimTime.Yearid, dimVenue.venue_name

    UNION ALL

    SELECT
        0 AS drug_treatment,
        0 AS QuarantineTotal,
        COUNT(FactVaccine.vaccineid) AS VaccineCount,
        COALESCE(dimTime.Yearid, 'UNKNOWN') AS Year,
        COALESCE(dimVenue.venue_name, 'UNKNOWN') AS venue_name,
        COALESCE(dimVaccine.vaccinename, 'UNKNOWN') AS vaccine_name
    FROM FactVaccine
    LEFT JOIN dimCitizen ON FactVaccine.citizenid = dimCitizen.citizenid
    LEFT JOIN dimTime ON FactVaccine.dateid = dimTime.dateid
    LEFT JOIN dimVenue ON FactVaccine.venueid = dimVenue.venueid
    LEFT JOIN dimVaccine ON FactVaccine.vaccineid = dimVaccine.vaccineid
    GROUP BY dimTime.Yearid, dimVenue.venue_name,dimVaccine.vaccinename

    UNION ALL

    SELECT
        COALESCE(SUM(FactDrug.dose), 0) AS drug_treatment,
        0 AS QuarantineTotal,
        0 AS VaccineCount,
        COALESCE(dimTime.Yearid, 'UNKNOWN') AS Year,
        COALESCE(dimVenue.venue_name, 'UNKNOWN') AS venue_name,
        'NONE' AS vaccine_name
    FROM FactDrug
    LEFT JOIN dimCitizen ON FactDrug.patientid = dimCitizen.citizenid
    LEFT JOIN dimTime ON FactDrug.dateid = dimTime.dateid
    LEFT JOIN dimVenue ON FactDrug.venueid = dimVenue.venueid
    GROUP BY dimTime.Yearid, dimVenue.venue_name


) AS Combined
GROUP BY Year, venue_name,vaccine_name;

''')
```

```python
#Result
m1= pd.read_sql('SELECT * FROM M1_Q1_Q2', conn)
m1
```

```python
M1 = pd.read_sql('''
SELECT
    Year,
    venue_name,
    vaccine_name,
    SUM(drug_treatment) AS drug_treatment,
    SUM(QuarantineTotal) AS QuarantineTotal,
    SUM(VaccineCount) AS VaccineCount,
    SUM(PatientCount) AS PatientCount,
    patientstatusname
FROM (
    SELECT
        0 AS drug_treatment,
        COUNT(FactQuarantine.citizenid) AS QuarantineTotal,
        0 AS VaccineCount,
        0 AS PatientCount,
        COALESCE(dimTime.Yearid, 'UNKNOWN') AS Year,
        COALESCE(dimVenue.venue_name, 'UNKNOWN') AS venue_name,
        'NONE' AS vaccine_name,
        NULL AS patientstatusname
    FROM FactQuarantine
    LEFT JOIN dimCitizen ON FactQuarantine.citizenid = dimCitizen.citizenid
    LEFT JOIN dimTime ON FactQuarantine.dateid = dimTime.dateid
    LEFT JOIN dimVenue ON FactQuarantine.venueid = dimVenue.venueid
    GROUP BY dimTime.Yearid, dimVenue.venue_name

    UNION ALL

    SELECT
        0 AS drug_treatment,
        0 AS QuarantineTotal,
        COUNT(FactVaccine.vaccineid) AS VaccineCount,
        0 AS PatientCount,
        COALESCE(dimTime.Yearid, 'UNKNOWN') AS Year,
        COALESCE(dimVenue.venue_name, 'UNKNOWN') AS venue_name,
        COALESCE(dimVaccine.vaccinename, 'UNKNOWN') AS vaccine_name,
        NULL AS patientstatusname
    FROM FactVaccine
    LEFT JOIN dimCitizen ON FactVaccine.citizenid = dimCitizen.citizenid
    LEFT JOIN dimTime ON FactVaccine.dateid = dimTime.dateid
    LEFT JOIN dimVenue ON FactVaccine.venueid = dimVenue.venueid
    LEFT JOIN dimVaccine ON FactVaccine.vaccineid = dimVaccine.vaccineid
    GROUP BY dimTime.Yearid, dimVenue.venue_name,dimVaccine.vaccinename

    UNION ALL

    SELECT
        COALESCE(SUM(FactDrug.dose), 0) AS drug_treatment,
        0 AS QuarantineTotal,
        0 AS VaccineCount,
        0 AS PatientCount,
        COALESCE(dimTime.Yearid, 'UNKNOWN') AS Year,
        COALESCE(dimVenue.venue_name, 'UNKNOWN') AS venue_name,
        'NONE' AS vaccine_name,
        NULL AS patientstatusname
    FROM FactDrug
    LEFT JOIN dimCitizen ON FactDrug.patientid = dimCitizen.citizenid
    LEFT JOIN dimTime ON FactDrug.dateid = dimTime.dateid
    LEFT JOIN dimVenue ON FactDrug.venueid = dimVenue.venueid
    GROUP BY dimTime.Yearid, dimVenue.venue_name

    UNION ALL

    SELECT
        0 AS drug_treatment,
        0 AS QuarantineTotal,
        0 AS VaccineCount,
        COUNT(*) AS PatientStatus,
        COALESCE(dimTime.Yearid, 'UNKNOWN') AS Year,
        COALESCE(dimVenue.venue_name, 'UNKNOWN') AS venue_name,
        COALESCE(dimPatientstatus.patientstatusname, 'UNKNOWN') AS patientstatusname,
        'NONE' AS vaccine_name
    FROM FactPatient
    LEFT JOIN dimCitizen ON FactPatient.patient = dimCitizen.citizenid
    LEFT JOIN dimTime ON FactPatient.dateid = dimTime.dateid
    LEFT JOIN dimVenue ON FactPatient.venueid = dimVenue.venueid
    LEFT JOIN dimPatientstatus ON FactPatient.patientstatusid = dimPatientstatus.patientstatusid
    WHERE FactPatient.patientstatusid = 1
    GROUP BY dimTime.Yearid, dimVenue.venue_name, dimPatientstatus.patientstatusname

) AS Combined
GROUP BY Year, venue_name,vaccine_name;

''', conn)
M1
```
